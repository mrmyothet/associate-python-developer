```python
import random

def place_O(board):
    """
    Given a 3x3 Tic-tac-toe board (list of lists),
    randomly choose an empty spot (0) and place a 2 (O).
    Returns the updated board.
    """
    empty_positions = []
    
    # Collect empty positions using loops (no list comprehension)
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                empty_positions.append((i, j))
    
    if len(empty_positions) == 0:
        raise ValueError("No empty spots available on the board.")
    
    # Randomly select one empty position
    pos = random.choice(empty_positions)
    i, j = pos
    board[i][j] = 2
    
    return board
```

```python
board = [
    [1, 0, 2],
    [0, 1, 0],
    [0, 0, 1]
]

new_board = place_O(board)
for row in new_board:
    print(row)
```

```python
board = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]

new_board = place_O(board)
for row in new_board:
    print(row)
```
