```python
def modify(num_list):
    num_list[1] = 99

    print("Inside function:", num_list)


my_list = [10, 20, 30]
modify(my_list)

print("Outside function:", my_list)
```

```python
my_list = [10, 20, 30]
modify(my_list.copy())

print("Outside function:", my_list)
```

```python
def modify(num_list):
    num_list[2][0] = 99

    print("Inside function:", num_list)


my_list = [10, 20, [30, 40]]
modify(my_list.copy())

print("Outside function:", my_list)
```
