```python
# Program to print a diamond of asterisks

# Ask the user for the height of the diamond (top half)
height = int(input("Enter the height of the diamond: "))

# Top half of the diamond
for i in range(1, height + 1):
    # Print spaces then stars
    print(" " * (height - i) + "*" * (2 * i - 1))

# Bottom half of the diamond
for i in range(height - 1, 0, -1):
    # Print spaces then stars
    print(" " * (height - i) + "*" * (2 * i - 1))
```

```python
# Program to print a giant letter A

# Ask the user for the size (height of the A)
size = int(input("Enter the size of the letter A: "))

for i in range(size):
    if i == 0:
        # Top of the A (single star centered)
        print(" " * (size - i - 1) + "*")
    elif i == 1:
        # Second row (two stars with a space between)
        print(" " * (size - i - 1) + "* *")
    elif i == 2:
        # Middle bar of the A (all stars)
        print(" " * (size - i - 1) + "*" * (2 * i + 1))
    else:
        # Lower rows (two stars with spaces in between)
        print(" " * (size - i - 1) + "*" + " " * (2 * i - 1) + "*")
        
```

```python
# Number guessing game where the computer guesses

print("Think of a number between 1 and 100, and I will try to guess it!")
print("After each guess, type: 'too high', 'too low', or 'correct'.")

low = 1
high = 100
found = False

while not found and low <= high:
    guess = (low + high) // 2   # computer guesses the midpoint
    print(f"My guess is: {guess}")
    feedback = input("Your feedback: ").strip().lower()

    if feedback == "too high":
        high = guess - 1
    elif feedback == "too low":
        low = guess + 1
    elif feedback == "correct":
        print("Yay! I guessed your number!")
        found = True
    else:
        print("Please respond with 'too high', 'too low', or 'correct'.")

if not found:
    print("Hmm, something went wrong. Did you change your number?")
```

```python
import random

my_number = random.randint(1, 100)

guess_number = input("Guess a number between 1 and 100:")
guess_number = int(guess_number)

counter = 1

while guess_number != my_number and counter < 10:
    if guess_number > my_number:
        print("Too high!")
    else:
        print("Too low!")
    guess_number = int(input("Guess again:"))
    counter += 1
```
