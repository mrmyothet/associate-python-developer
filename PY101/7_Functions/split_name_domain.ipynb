```python
def split_name_domain(email):
    """Splits an email address into its name and domain parts.

    Args:
        email (str): The email address to split.
    Returns:
        No return value. Prints the name and domain parts.
    """

    email_list = email.split("@")
    name = email_list[0]
    domain = email_list[1]
    print("Name: {} \nDomain: {}".format(name, domain))
```

```python
split_name_domain("rinchen@gmail.com")
split_name_domain("karma@gmail.com")
split_name_domain("kiran@gmail.com")
```
