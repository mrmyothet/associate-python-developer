```python

def validate_user(name:str, email:str, password: str):
    """
    This function validate the user details to register to the app.

    Args: 
      name (str): user name
      email (str): email
      password (str): password

    Return: 
      result (boolean): whether pass or fail
    """
    
    if validate_name(name) and validate_email(email) and validate_password(password): 
        return True
    else: 
        raise ValueError("validate_user function expects name, email and password arguments as string")
```
